```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
y = pd.DataFrame(iris['target'], columns=['target_names'])
```

```python
x
```

```python
y
```

```python
data = pd.concat([x,y], axis=1)
data
```

```python
a = np.arange(0, 150)
i = data['sepal length (cm)']
j = data['sepal width (cm)']
k = data['petal length (cm)']
l = data['petal width (cm)']
```

```python
plt.scatter(a, i)
plt.xlabel('number')
plt.ylabel('sepal length (cm)')
```

```python
plt.scatter(a, j)
plt.xlabel('number')
plt.ylabel('sepal width (cm)')
```

```python
plt.scatter(i, j, c='g')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
o = np.arange(0,150)
plt.scatter(i, j, c=o)
```

```python
w = np.array(data['target_names'])
w
plt.scatter(i, j, c=w)
```

```python
plt.scatter(a, i/j)
plt.xlabel('number')
plt.ylabel('sepal length / sepal width')
```

```python
plt.scatter(a, k)
plt.xlabel('number')
plt.ylabel('petal length (cm)')
```

```python
plt.scatter(a, l)
plt.xlabel('number')
plt.ylabel('petal width (cm)')
```

```python
plt.scatter(k, l)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
plt.scatter(a, k/l)
plt.xlabel('number')
plt.ylabel('petal length / petal width')
```
